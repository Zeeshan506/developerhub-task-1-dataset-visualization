```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

Step 2: Load the Dataset

```python
iris = sns.load_dataset("iris")
# since we are not using a local csv file we cant use pandas to load it, we will directly load it via seaborn library
# if we were to load the dataset locally here is how it would be done
# iris = pd.read_csv("Path_to_csv/(CSV_name)")
```

Step 4: Use .info() and .describe() for summary statistics.

```python
print("-----------------info-----------------\n")
print(iris.info())

print("\n-----------------Statistic Summary-----------------\n")
print(iris.describe())

print("\n-----------------Shape-----------------\n")
print(f"Rows: {iris.shape[0]}")
print(f"Columns: {iris.shape[1]}")


```

### ***Visualizing the dataset:***

Create a scatter plot to show relationships between features

```python
# Set up 4 subplots in 1 row
fig, axes = plt.subplots(1, 4, figsize=(20, 5))  # (rows, cols), adjust figsize as needed

# Plot 1: Petal Scatter
sns.scatterplot(data=iris, hue="species", x='petal_width', y='petal_length', ax=axes[0])
axes[0].set_title("Petal Scatter")

# Plot 2: Sepal Scatter
sns.scatterplot(data=iris, hue="species", x='sepal_width', y='sepal_length', ax=axes[1])
axes[1].set_title("Sepal Scatter")

# Plot 3: Width Scatter
sns.scatterplot(data=iris, hue="species", x='sepal_width', y='petal_width', ax=axes[2])
axes[2].set_title("Width Scatter")

# Plot 4: Length Scatter
sns.scatterplot(data=iris, hue="species", x='sepal_length', y='petal_length', ax=axes[3])
axes[3].set_title("Length Scatter")

plt.suptitle("Scatar Plot for Relation or Corelation ", fontsize=16)
plt.tight_layout()
plt.show()
```

Use histograms to show value distributions.

```python
fig, axes = plt.subplots(1, 4, figsize=(20,4))
# follow the same syntax as done before, just change the name and the feature of the plot.
#  v-> change here
sns.histplot(data=iris, x='sepal_width', kde = True, ax = axes[0])
axes[0].set_title("Sepal Width")

sns.histplot(data=iris, x='sepal_length', kde = True, ax=axes[1])
axes[1].set_title("Sepal Length")

sns.histplot(data=iris, x='petal_width', kde = True,  ax=axes[2])
axes[2].set_title("Petal Width")

sns.histplot(data=iris, x='petal_length', kde = True, ax=axes[3])
axes[3].set_title("Petal Length")

plt.suptitle("Histplot for Value Distribution", fontsize=16)
plt.tight_layout()
plt.show()
```

Use box plots to identify outliers.

```python
fig, axes = plt.subplots(1, 4, figsize=(20,4))

sns.boxplot(data=iris, x='sepal_width', ax = axes[0])
axes[0].set_title("Sepal Width")

sns.boxplot(data=iris, x='sepal_length', ax=axes[1])
axes[1].set_title("Sepal Length")

sns.boxplot(data=iris, x='petal_width',  ax=axes[2])
axes[2].set_title("Petal Width")

sns.boxplot(data=iris, x='petal_length', ax=axes[3])
axes[3].set_title("Petal Length")

plt.suptitle("Box Plots for Outlier Detection", fontsize=16)
plt.tight_layout()
plt.show()
```

## ***Summary***:
- Iris dataset has 150 rows and 4 numerical features + 1 label or target.
- Species are fairly well-separated in petal features (esp. petal length vs petal width).
- Some overlap exists in sepal features. esp in sepal width vs sepal length. the plot is cloudy indicating overlaps.
- Boxplots show slight outliers in sepal width. to be consize 3.
- Histograms confirm most values are normally distributed across the dataset
